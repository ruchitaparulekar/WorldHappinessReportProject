### Exploring the data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
data = pd.read_csv("tested.csv")
data.head(10)
```

```python
data.columns
```

```python
data.describe()
```

```python
data.isnull().sum()
```

We can observe that there are the following columns have null vales: Cabin, Age and Fare. We can fill in the remaining age values by using forwardfill, as out 419 only 86 values are missing.

```python
data['Age'].fillna(method='ffill',inplace=True)
data.head(20)
```
